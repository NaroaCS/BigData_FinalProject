```python
# import
import folium
import geojson
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely import wkt
```

```python
# Open data files

park = gpd.read_file('data/park_properties.geojson')

parkaccess = gpd.read_file('data/park_access.json')
internet = gpd.read_file('data/internet.geojson')
food = pd.read_csv('data/snap.csv')
```

```python
#combine data

combined_data = parkaccess
internet['boro_ct2010']=internet['boro_ct2010'].astype(int)
food.boro_ct2010.astype(int)
combined_data = combined_data.merge(internet[['boro_ct2010', 'PercentageU18WithAccess']], on='boro_ct2010')
combined_data = combined_data.merge(food, on='boro_ct2010')
combined_data['snap_perc'] = combined_data['snap_perc'].astype(float)
```
