```python
import pandas as pd
raw_data = pd.read_csv("./InternetComputerAge.csv") #importing CSV to DF
print(raw_data.head())
print(raw_data.columns) # This has a lot of columns! Take a look at the data dictionary "R12438492.txt" and think about what columns you will need for this project
```

```python
#YOUR CODE HERE
cleaned_data =raw_data.drop(raw_data.columns.difference(['GEO_ID', 'B28005_002E','B28005_005E']), 1)
print(type(cleaned_data))
```

```python
#steps: a) call DataFrame.head() to check out what your data looks like:
#         hint: is the first row also data you need? remove the first row if you don't need them
#YOUR CODE HERE
#celeaned_data['Percentage']

cleaned_data.drop([0, 0], inplace=True)
cleaned_data.rename({'B28005_002E': 'TotalUnder18', 'B28005_005E': 'WithComputerAndBroadband'}, axis=1, inplace=True)
cleaned_data = cleaned_data.reset_index(drop=True)
print(cleaned_data.head())
```

```python
# Step b) you may want to reset_index() so that your index begins with 0 if you remove any rows
#YOUR CODE HERE

print(type(cleaned_data))
cleaned_data.dtypes
cleaned_data_numeric = cleaned_data.apply(pd.to_numeric, errors='ignore')
cleaned_data_numeric.dtypes
```

```python
# Step c)call DataFrame.head() again to make sure your new index is saved in the memory
#YOUR CODE HERE
cleaned_data_numeric['PercAcess']=(cleaned_data_numeric['WithComputerAndBroadband']/cleaned_data_numeric['TotalUnder18'])*100
cleaned_data_numeric.head()
```

```python
#YOUR CODE HERE
import math

cleaned_data_numeric['GEO_ID'].dtypes

for i in range(len(cleaned_data_numeric)):
    
    county_id=cleaned_data_numeric['GEO_ID'][i][11:14]
    censustract_id=cleaned_data_numeric['GEO_ID'][i][14:]

    if county_id =='061': #NY County
        boro_id='1' #Manhattan
    elif county_id == '005': #Bronx County
        boro_id= '2'#The Bronx
    elif county_id=='047': #King's County
        boro_id='3'#Brooklyn
    elif county_id=='081':#Queens County
        boro_id='4' #Queens
    elif county_id=='085': #Richmond County
        boro_id='5' #Staten Island
    else:
        boro_id='ERROR'

    cleaned_data_numeric['GEO_ID'][i]=boro_id+censustract_id
    
    if math.isnan(cleaned_data_numeric['PercAcess'][i]):
        cleaned_data_numeric['PercAcess'][i]=0
        print(i, cleaned_data_numeric['PercAcess'][i])
    
print(cleaned_data_numeric.head())
    
cleaned_data_numeric.to_csv('InternetAccessData.csv', index = False)
```

```python
import json

with open('CT2010.geojson','r') as f:
    ct_geojson = json.load(f)


internet_dict=cleaned_data_numeric.set_index('GEO_ID')['PercAcess']
#print(internet_dict[50009001000])

#Loop over GeoJSON features and add the new properties
for feat in ct_geojson['features']:
    try: 
        ct_id=feat['properties']['boro_ct2010']
        perc=internet_dict[ct_id]
        feat['properties']['PercentageU18Access']=int(perc)
    
    except KeyError:
        pass

#Write result to a new file
with open('new_str.geojson', 'w') as f:
    json.dump(ct_geojson, f)

    
```
